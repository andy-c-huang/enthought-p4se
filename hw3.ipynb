```python
"""
Calculate Derivative
--------------------

Topics: NumPy array indexing and array math.

Use array slicing and math operations to calculate the
numerical derivative of ``sin`` from 0 to ``2*pi``.  There is no
need to use a 'for' loop for this.

Plot the resulting values and compare to ``cos``.

Bonus
~~~~~

Implement integration of the same function using Riemann sums or the
trapezoidal rule.

See :ref:`calc-derivative-solution`.
"""
from numpy import linspace, pi, sin, cos, cumsum
import matplotlib.pyplot as plt
# calculate the sin() function on evenly spaced data.
x = linspace(0,2*pi,101)
y = sin(x)

plt.plot(x,y)
plt.show()
```

```python
z = (y[1:]-y[:-1])/(x[1:]-x[:-1])
x_mid = (x[1:]+x[:-1])/2.0
plt.plot(x_mid,z)
plt.show()
```

```python
int_z = z.cumsum
plt.plot(x_mid,int_z())
plt.show()
```

```python
import numpy
z = numpy.zeros(img.shape)
l1 = z.copy() ; l1[:,:-1] = img[:,1:]
r1 = z.copy() ; r1[:,1:] = img[:,:-1]
u1 = z.copy() ; u1[:-1,:] = img[1:,:]
d1 = z.copy() ; d1[1:,:] = img[:-1,:]

plt.imshow((img + l1 + r1 + u1 + d1)/5, cmap=plt.cm.hot)
plt.show()
```

```python
plt.imshow((img + l1 + r1 + u1 + d1)/5 - img, cmap=plt.cm.hot)
plt.show()
```

```python
def l_shift_n(img,n):
  l = numpy.zeros(img.shape) ; l[:,:-n] = img[:,n:]
  return l

def r_shift_n(img,n):
  r = numpy.zeros(img.shape) ; r[:,n:] = img[:,:-n]
  return r

def u_shift_n(img,n):
  u = numpy.zeros(img.shape) ; u[:-n,:] = img[n:,:]
  return u

def d_shift_n(img,n):
  d = numpy.zeros(img.shape) ; d[n:,:] = img[:-n,:]
  return d

def blur_5pt(img):
  b = (img + l_shift_n(img,1) + r_shift_n(img,1) + u_shift_n(img,1) + d_shift_n(img,1))/5
  return b

plt.close()

p = plt.subplots(nrows=1, ncols=5,figsize=(10, 5), dpi=100)
figures = p[1]

figures[0].imshow(l_shift_n(img,50), cmap=plt.cm.hot)
figures[0].set_title('L50')

figures[1].imshow(r_shift_n(img,50), cmap=plt.cm.hot)
figures[1].set_title('R50')

figures[2].imshow(u_shift_n(img,50), cmap=plt.cm.hot)
figures[2].set_title('U50')

figures[3].imshow(d_shift_n(img,50), cmap=plt.cm.hot)
figures[3].set_title('D50')

figures[4].imshow(blur_5pt(img), cmap=plt.cm.hot)
figures[4].set_title('Blur5pt')


plt.show()
```
